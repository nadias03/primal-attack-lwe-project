```python
SEEDS = [i for i in range(10)]
```

```python
easy_results = []

for seed in SEEDS:
    print("================================================")
    print(f"Easy attack demonstration for seed {seed}")
    np.random.seed(seed=seed)

    easy_res = run_primal_attack(
        n=4,
        m=8,
        q=97,
        S=[-1, 0, 1],
        E=[-1, 0, 1],
    )

    easy_results.append(easy_res)
    print("================================================")
```

```python
# jeszcze trudniejszy przykład
result_medium = run_primal_attack(
    reduction_mathod="LLL",
    n=24,
    m=48,
    q=769,
    S=[-2, -1, 0, 1, 2],
    E=[-3, -2, -1, 0, 1, 2, 3],
)
```

```python
medium_results = []

for seed in SEEDS:
    print("================================================")
    print(f"Medium attack demonstration for seed {seed}")
    np.random.seed(seed=seed)

    res_medium = run_primal_attack(
        reduction_mathod="LLL",
        n=24,
        m=48,
        q=769,
        S=[-2, -1, 0, 1, 2],
        E=[-3, -2, -1, 0, 1, 2, 3],
    )

    medium_results.append(res_medium)
    print("================================================")
```
